```python
from pymongo import MongoClient
from tqdm import tqdm
import orjson
import os
```

```python
tier = 'diamond'
data_dir = '/home/data/lol/matches/gold/'

client = MongoClient("mongodb://localhost:27017/")
db = client["loldb"]
collection_match = db[f'{tier}_match']
```

```python
list(collection_match.find().limit(1))
```

```python
duplicates = collection_match.aggregate([
  {
    '$group': {
      '_id': "$id",
      'uniqueIds': { '$addToSet': "$_id" },
      'count': { '$sum': 1 }
    }
  },
  {
    '$match': {
      'count': { '$gt': 1 }
    }
  }
])

for doc in tqdm(duplicates):
    doc['uniqueIds'].pop(0) # 첫 번째 _id를 유지하고 나머지를 삭제하기 위해 첫 번째 요소를 제거
    if doc['uniqueIds']: # 삭제할 문서가 있을 경우만 실행
        collection_match.delete_many({
            "_id": { "$in": doc['uniqueIds'] } # 나머지 _id를 가진 문서들을 삭제
        })
```

```python
from pymongo import MongoClient
import re

# MongoDB 데이터베이스에 연결합니다.
client = MongoClient('mongodb://localhost:27017/')
db = client['loldb']  # 실제 데이터베이스 이름으로 대체하세요.

# 'match_'로 시작하는 컬렉션 이름을 찾기 위한 정규식 패턴입니다.
pattern = re.compile("^riot_.*_to_index$")

# 데이터베이스의 모든 컬렉션 이름을 검색합니다.
collection_names = db.list_collection_names()

# 정규식과 일치하는 컬렉션을 찾아서 삭제합니다.
for name in collection_names:
    if pattern.match(name):
        # db.drop_collection(name)
        print(f'Dropped collection: {name}')
```
